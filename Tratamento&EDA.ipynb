```python
import pandas as pd
pd.set_option('display.max_columns', None)

import utils as ut
```

```python
# Criando dataframes

urls_convertidas = [ut.conversor_url(url) for url in ut.urls] # convertendo urls de compartilhamento do sheets

BC_ES_bruto = pd.read_csv(urls_convertidas[0],parse_dates=['Data/Hora'],dayfirst=True,low_memory=False)[ut.colunas_selecionadas]
BS_SC_PR_bruto = pd.read_csv(urls_convertidas[1],parse_dates=['Data/Hora'],dayfirst=True,low_memory=False)[ut.colunas_selecionadas]
BS_SP_bruto = pd.read_csv(urls_convertidas[2],parse_dates=['Data/Hora'],dayfirst=True,low_memory=False)[ut.colunas_selecionadas]
BS_RJ_bruto = pd.read_csv(urls_convertidas[3],parse_dates=['Data/Hora'],dayfirst=True,low_memory=False)[ut.colunas_selecionadas]
RN_CE_bruto = pd.read_csv(urls_convertidas[4],parse_dates=['Data/Hora'],dayfirst=True,low_memory=False)[ut.colunas_selecionadas]
SE_AL_bruto = pd.read_csv(urls_convertidas[5],parse_dates=['Data/Hora'],dayfirst=True,low_memory=False)[ut.colunas_selecionadas]
```

```python
BC_ES_bruto.head() # Exemplo de dataframe criado
```

```python
# Armazenando dataframes em dicionário

dataframes = {'BC_ES': BC_ES_bruto,
 'BS_SC_PR': BS_SC_PR_bruto,
 'BS_SP': BS_SP_bruto,
 'BS_RJ': BS_RJ_bruto,
 'RN_CE': RN_CE_bruto,
 'SE_AL': SE_AL_bruto
}
```

```python
# Checando tamanho dos dataframes

for nome, df in dataframes.items():
    print(f'{nome}: {df.shape}')
```

```python
# Checando duplicatas

for nome, df in dataframes.items():
    print(f'{nome}: {df.duplicated().sum()} registros duplicados')
```

```python
# Checando nulos

for nome, df in dataframes.items():
    print(nome)
    print(df.isna().sum()[df.isna().sum() > 0])
    print('-' * 33)
```
